 <h3>Default credit card dataset</h3>



<b>Dataset Information</b>

This dataset contains information on default payments, demographic factors, credit data, history of payment, and bill statements of credit card clients in Taiwan from April 2005 to September 2005.

<b>The Business Context</b>

A Taiwan-based credit card issuer wants to better predict the likelihood of default for its customers, as well as identify the key drivers that determine this likelihood. This would inform the issuer’s decisions on whom to give/issue a credit card to and what credit limit to be provided. It would also help the issuer to have a better understanding of their current and potential customers, which would inform their future strategy, including their planning of offering targeted credit products to their customers.

This advanced level data set has <b>30000 rows and 25 columns</b>.
The data set could be used to estimate the probability of default payment by credit card client using the data provided.

This data set is recommended for learning and practicing your skills in <b> exploratory data analysis</b>, <b>data visualization, and classification modelling techniques. </b>Feel free to explore the data set with multiple supervised and unsupervised learning techniques.

```python
dupe_ids = id_counts.index[dupe_mask] #storing all the duplicate index into new variable and converting into list and finding length of the list.
dupe_ids = list(dupe_ids)
len(dupe_ids)
```

```python
dupe_ids[0:5] #printing upto 5 duplicate ids
```

```python
df.loc[df['ID'].isin(dupe_ids[0:3]),:].head(10) #finding location of 1st 3 dup ids among 313.
```

We can see some duplicates here, and it looks like every duplicate ID has one row with data, and another row with all zeros. Is this the case for every duplicate ID? Let's check.

```python
df.shape
```

```python
df_zero_mask = df == 0
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis=1) # 
sum(feature_zero_mask)

#iloc used to select particular cell of the dataset. Here [] row(series) ,[[]] r & c (data frame)
```


It looks like there are at least as many "zero rows" as there are duplicate IDs. Let's remove all the rows with all zero features and response, and see if that gets rid of the duplicate IDs.

```python
df_clean_1 = df.loc[~feature_zero_mask,:].copy() #removed all the duplicates and stored in new variable.

df_clean_1.shape
```

```python
df_clean_1['ID'].nunique() #contains only the unique values.
```

```python
from scipy import stats
import numpy as np
#Generate some synthetic near-linear data:
```

```python
X = np.array(range(1,11))
X #here we are creating an array with given range of elements.
```

```python
np.random.seed(seed=24)
Y = 2*X + np.random.normal(0, 1, 10) #low,high,size
```

```python
slope_line, intercept_line, r_value, p_value, std_err = stats.linregress(X,Y)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 400 
mpl.rcParams['font.size'] = 16
plt.plot(X, intercept_line + slope_line*X, 'r-') #
plt.bar(X,Y, align='center')
plt.xlabel('Self-reported customer satisfaction on scale from 1-10') #label for X-axis
plt.ylabel('Average time spent on\nwebsite, example 1 (minutes)') #label for y-axis
plt.yticks(range(0,25,5)) # y-axis scale(range is from 0 upto 25 with 5units difference )
```

And some synthetic polynomial data

```python
np.random.seed(seed=20)
noise = np.random.normal(0, 1, 10)
Y_poly = -1*(X-2)*(X-9) + 10 + noise
slope_poly, intercept_poly, r_value, p_value, std_err = stats.linregress(X,Y_poly)
```

```python
plt.plot(X, intercept_poly + slope_poly*X, 'r-')
plt.bar(X,Y_poly,align='center')
plt.xlabel('Self-reported customer satisfaction on scale from 1-10')
plt.ylabel('Average time spent on\nwebsite, example 2 (minutes)')
```
